```python
import pandas as pd
import xarray as xr
import numpy as np
```

```python
# simple multiplication table
x = np.arange(1,10)
y = np.arange(1,10)
X,Y = np.meshgrid(x,y)
data = X * Y
```

```python
df = pd.DataFrame(data, index=y, columns=x)
xs = xr.Dataset.from_dataframe(df)
```

```python
# Create Dataset
ds = xr.Dataset(
    data_vars=dict(
        data=(["x", "y"], data),
    ),
    coords=dict(
        x=x,
        y=y,
    ),
    attrs=dict(description="Multiplication Table"),
)
ds
```

```python
xr.DataArray
```

```python
# Create DataArray
da = xr.DataArray(
    data=data,
    dims=["x", "y"],
    coords=dict(
        x=x,
        y=y,
    ),
    attrs=dict(
        description="Multiplication table",
    )
)
da
```

```python
# stack several dimensions into single one
da.stack(flat=('x', 'y'))
```

```python
#select using ranges
da.loc[ 5:10, 1:5 ]
```

```python
# locate using single values
da[dict(x=1)]
```
