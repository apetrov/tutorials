```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
```

### Problem

Calculate approximate posterior distribution given prior distribution and data
Toy Example of what's going on under the hood of PyMC3

```python
x = np.linspace(0, 1)
n = 10
k = 3
prior = scipy.stats.beta(10,10).pdf(x)
likelihood = scipy.stats.binom(n, x).pmf(k)
posterior_kernel =   prior * likelihood
dx = x[1] - x[0]
posterior = posterior_kernel/np.sum(posterior_kernel * dx)
```

```python
plt.plot(x, prior, label='prior')
plt.plot(x, likelihood, label='likelihood')
plt.plot(x, posterior, label='posteior')
plt.legend()
```

```python
prior_log = np.log(scipy.stats.beta(10,10).pdf(x) + 1e-10)
likelihood_log = np.log(scipy.stats.binom(n, x).pmf(k) + 1e-10)
posterior_log =   prior_log + likelihood_log
```

```python
plt.plot(x, prior_log, label='prior_log')
plt.plot(x, likelihood_log, label='likelihood_log')
plt.plot(x, posterior_log, label='posteior_log')
plt.legend()
```
