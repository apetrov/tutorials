```python
import numpy as np
import pandas as pd
import scipy.stats
import seaborn as sns
```

```python
y = np.linspace(0.5,3) #  rows
x = np.linspace(1,5) # columns
```

```python
X,Y = np.meshgrid(x,y)
```

```python
data = scipy.stats.norm(Y, 1).pdf(X)
```

```python
df = pd.DataFrame(data, columns=x, index=y)
sns.heatmap(df)
```
